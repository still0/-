```python
import threading 
import requests
from requests.exceptions import RequestException
from pyquery import PyQuery as pq
import time
import csv
requests.packages.urllib3.disable_warnings(requests.packages.urllib3.exceptions.InsecureRequestWarning)  #禁止提醒SSL警告

class ip_test():
    
    def __init__(self):
        self.goodip = []
        self.iplist = self.read_ip_list()
        self.listlen = len(self.iplist)        
        self.threadmax = threading.BoundedSemaphore(20)
        
    def a_iptest(self,proxy_test):
          
        url = 'https://www.tmall.com/'#'https://www.baidu.com/'
        
        proxies = {'http': 'http://' + proxy_test,'https': 'https://' + proxy_test,}
        try:
            response0 = requests.get(url, proxies=proxies,verify = False,timeout = 5)
                    
            if response0.status_code == 200:
                self.goodip.append([proxy_test,str(response0.elapsed.microseconds/1000)])
                print(proxy_test + ' 测试OK，所用时间'+str(response0.elapsed.microseconds/1000) + 'ms')
        except:
            print(proxy_test + ' 测试失败')
            pass
    
    def read_ip_list(self):
        self.proxy_tests = []
        with open('goodip.csv','r',encoding = 'GBK') as csvfile2:
            reader = csv.reader(csvfile2)            
            for row in reader:
                self.proxy_tests.append(row[0])
            return self.proxy_tests
       
    def main(self, num = 0):
        with self.threadmax:
            ip = self.iplist[num]
            self.a_iptest(ip)
        
        
        
if __name__ == '__main__' :
    iptest = ip_test() #创建对象getiplist
    
    
    threads = []
    #print(iptest.iplist)
    #iptest.main(1) 测试主函数可否正常运行
    
    for i in range(iptest.listlen):#
        t = threading.Thread(target = iptest.main,args = (i,) )
        threads.append(t)
        
    for t in threads:
        t.setDaemon(True)
        t.start()
    for t in threads:
        t.join()
    print(iptest.goodip)
    with open('baidugoodip.csv','w',newline = '') as csvfile:
        writer = csv.writer(csvfile)
        for ipinfo in iptest.goodip:
            writer.writerow(ipinfo)
        print('已成功保存csv文件')
                            
            
            
            
    print("可以使用的iP数: " + str(len(iptest.goodip)))
  
```
